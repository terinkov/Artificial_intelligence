```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')
sns.set(font_scale=1.3)
irises = pd.read_excel('iris_data.xlsx')
irises.head()
```

```python
type(irises['Длина чашелистника'][0])
```

```python
irises.describe()
iries.append([10,0.7,2,1.1,'Iris-virginica'])
```

```python
 irises.loc[-1] = [10,0.7,2,1.1,'Iris-virginica']  # adding a row
 irises.index = irises.index + 1  # shifting index
 df = irises.sort_index()  # sorting by index
```

```python
irises.describe()
```

```python
g = sns.pairplot(irises, hue = 'Цветок', plot_kws = {'alpha': 0.9, 's': 80})
g.map(sns.regplot)
g.add_legend()

g
```

```python
g = sns.pairplot(irises, hue = 'Цветок', plot_kws = {'alpha': 0.6, 's': 80})
g.map(sns.regplot)
g.add_legend()

g
```

```python
# selected это все виды цветов
selected = irises.select_dtypes('float64')
```

```python
selected.corr()
```

```python
sns.heatmap(
    selected.corr(),        
    cmap='RdBu_r', # задаёт цветовую схему
    annot=True, # рисует значения внутри ячеек
    vmin=-1, vmax=1); # указывает начало цветовых кодов от -1 до 1.
```

```python
selected_Iris_setosa = irises[irises["Цветок"]=="Iris-setosa"].select_dtypes('float64')
sns.heatmap(
    selected_Iris_setosa.corr(),        
    cmap='RdBu_r', # задаёт цветовую схему
    annot=True, # рисует значения внутри ячеек
    vmin=-1, vmax=1); # указывает начало цветовых кодов от -1 до 1.
```

```python
selected_Iris_setosa = irises[irises["Цветок"]=="Iris-virginica"].select_dtypes('float64')
sns.heatmap(
    selected_Iris_setosa.corr(),        
    cmap='RdBu_r', # задаёт цветовую схему
    annot=True, # рисует значения внутри ячеек
    vmin=-1, vmax=1); # указывает начало цветовых кодов от -1 до 1.
```
